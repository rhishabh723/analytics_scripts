```python
query2=f"""
SELECT * FROM `perceptive-ivy-290216.f1_api.race_lap_time`
WHERE GP="Monaco Grand Prix"
AND YEAR=2024
"""
track2=pd.read_gbq(query2,project_id,dialect='standard')
```

```python
track2.head()
```

```python
track2["LapTime_final"]=track2['LapTime'].str.split('days ').str[1]
track2["LapTime"]=pd.to_timedelta(track2["LapTime"])
track2.loc[:, "LapTime (s)"] = track2["LapTime"].dt.total_seconds()
# track2['LapTime'] = track2['LapTime'].dt.time
# track2["LapTime"]=pd.to_datetime(track2['LapTime'],format='%H:%M:%S.%f')
# track2["LapTime"]=track2["LapTime"].dt.strftime("%H:%M:%S.%f")
track2.head()
```

```python
#Drop specific laps
track2 = track2[track2.LapNumber != 1]
```

```python
track2[track2["Driver"]=="VER"].sort_values(by=["LapNumber"])
```

```python
query3=f"""
SELECT * FROM `perceptive-ivy-290216.f1_api.race_lap_time`
WHERE GP="Bahrain Grand Prix"
# AND YEAR=2005
"""
track3=pd.read_gbq(query3,project_id,dialect='standard')
```

```python
track3['LapTime_TD']= pd.to_timedelta(track3["LapTime"])
track3["LapTime"]=track3['LapTime'].str.split('days ').str[1]
track3.loc[:, "LapTime (s)"] = track3["LapTime_TD"].dt.total_seconds()
track3.head()
```

```python
#Remove Lap Data that exceeds a certain value

track3=track3[track3["LapTime (s)"]<200.00]
```

```python
#Plot for a specific Lap
fig2=px.scatter(track3, x="Team", y="LapTime (s)",
                color='Team',
        # size='LapNumber',
        hover_data=["LapTime","Year","GP","Driver","Team","Compound","Position","LapStartTime","LapNumber","PitOutTime","PitInTime","FreshTyre","TyreLife","Sector1Time", "Sector2Time",'Sector3Time'],
        height=1000, width=1800,
        )
fig2.show()
```

```python
query4=f"""
SELECT * FROM `perceptive-ivy-290216.f1_api.results_qualifying`
WHERE GP="Bahrain Grand Prix"
# AND YEAR=2005
"""
track4=pd.read_gbq(query4,project_id,dialect='standard')
```

```python
track4['LapTime_TDQ1']= pd.to_timedelta(track4["Q1"])
track4["Q1"]=track4['Q1'].str.split('days ').str[1]
track4.loc[:, "LapTime Q1 (s)"] = track4["LapTime_TDQ1"].dt.total_seconds()

track4['LapTime_TDQ2']= pd.to_timedelta(track4["Q2"])
track4["Q2"]=track4['Q2'].str.split('days ').str[1]
track4.loc[:, "LapTime Q2 (s)"] = track4["LapTime_TDQ2"].dt.total_seconds()

track4['LapTime_TDQ3']= pd.to_timedelta(track4["Q3"])
track4["Q3"]=track4['Q3'].str.split('days ').str[1]
track4.loc[:, "LapTime Q3 (s)"] = track4["LapTime_TDQ3"].dt.total_seconds()

track4.head()
```
