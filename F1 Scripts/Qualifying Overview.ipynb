```python
query2=f"""
SELECT * FROM `perceptive-ivy-290216.f1_api.qualyfing_lap_time`
WHERE YEAR=2021
AND GP='Spanish Grand Prix'
"""
track2=pd.read_gbq(query2,project_id,dialect='standard')
```

```python
Drivers=(track2['Driver'].unique())
Drivers
```

```python
track2["LapTime"].min()
```

```python
#Assign Rank for each entry point
track2["RK"] = track2.groupby(by=["Driver"])["LapTime"].rank(method="dense", ascending=True)
track2["RK"]= track2["RK"].astype(int)
```

```python
track_fastest=track2[track2['RK']==1]
track_fastest
```

```python
pole_lap = track_fastest["LapTime"].min()
track_fastest['LapTimeDelta']= pd.to_timedelta(track_fastest["LapTime"]) - pd.to_timedelta(pole_lap)
track_fastest['LapTimeDelta'] = track_fastest['LapTimeDelta'] + pd.to_datetime('1970/01/01')

track_fastest['LapTimeDelta2']= pd.to_timedelta(track_fastest["LapTime"]) - pd.to_timedelta(pole_lap)
track_fastest['LapTimeDelta2']=track_fastest['LapTimeDelta2'].astype(str)

track_fastest["LapTimeDelta2"]=track_fastest['LapTimeDelta2'].str.split('days ').str[1]

# track_fastest['LapTimeDelta']=track_fastest['LapTimeDelta'].astype(str)
# for index, row in track_fastest.iterrows():
#   if track_fastest.loc[index, 'LapTimeDelta']=="0 days 00:00:00":
#     track_fastest.loc[index, 'LapTimeDelta']="0 days 00:00:00.000000"
# track_fastest["LapTimeDelta"]=track_fastest['LapTimeDelta'].str.split('days ').str[1]
# track_fastest["LapTimeDelta"]=pd.to_datetime(track_fastest['LapTimeDelta'],format='%H:%M:%S.%f')

# track_fastest['LapTime']=track_fastest['LapTime'].astype(str)
track_fastest.head()
```

```python
fig=px.bar(
    track_fastest,
    y="Driver",
    x='LapTimeDelta',
    text='LapTimeDelta2',
    orientation='h',
    color='Team',
    title="<b>Qualifying Results</b>",
    height=800, width=1200,

    color_discrete_map={
                 "Mercedes": "#6CD3BF",
                 "Red Bull Racing": "#3860C6",
                 "Alpine": "#2293D1",
                 "Mclaren": "#FFA500",
                 "Haas": "#B6BABD",
                 "Williams": "#37BEDD",
                 "Alfa Romeo Racing": "#C92D4B",
                 "Ferrari": "#F91536",
                 "Aston Martin": "#358C75",
                 "AlphaTauri": "#5E8FAA" }
)
fig.update_layout(
    yaxis = dict(tickfont = dict(size=20)),
    xaxis = dict(tickfont = dict(size=20)),

    xaxis_title="<b>Delta</b>",
    yaxis_title="<b>Driver</b>",
    # font_family="Courier New",
    # font_color="blue",
    title_font_family="<b>PT Sans Narrow</b>",
    # title_font_color="red",
    # legend_title_font_color="green"

#  "Arial", "Balto", "Courier New", "Droid Sans", "Droid Serif", "Droid Sans Mono", "Gravitas One", "Old Standard TT", "Open Sans", "Overpass", "PT Sans Narrow", "Raleway", "Times New Roman"
)


# fig['layout']['yaxis']['autorange'] = "reversed"
# fig['layout']['xaxis']['autorange'] = "reversed"
fig.update_layout(xaxis_tickformat='%H:%M:%S.%f')

fig.update_traces(marker_line_width=1,marker_line_color="BLACK")

fig.update_layout(yaxis={'categoryorder':'total descending'})

fig.update_traces(textposition='auto')

fig.show()
```
