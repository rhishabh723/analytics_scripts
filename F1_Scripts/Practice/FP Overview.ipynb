```python
df_bq_fp1=df_bq2_fp1[(df_bq2_fp1["GP"]==race)&(df_bq2_fp1["Year"]==2024)]
GP_fp1=df_bq_fp1
GP_fp1["LapNumber"]=GP_fp1["LapNumber"].astype(int)
GP_fp1=GP_fp1.sort_values(['Driver','LapNumber'])
GP_fp1["DriverNumber"]=GP_fp1["DriverNumber"].astype(int)
year=GP_fp1['Year'].iloc[0]
gp=GP_fp1['GP'].iloc[0]
GP_fp1.head()
```

```python
#Convert LapTime to Time Delta for plotting on Y Axis
GP_fp1['LapTime_TD']= pd.to_timedelta(GP_fp1["LapTime"])
GP_fp1["LapTime"]=GP_fp1['LapTime'].str.split('days ').str[1]
GP_fp1.loc[:, "LapTime (s)"] = GP_fp1["LapTime_TD"].dt.total_seconds()
```

# Fastest Driver in each lap

```python
GP_fp1['s1_TD']= pd.to_timedelta(GP_fp1["Sector1Time"])
GP_fp1['s2_TD']= pd.to_timedelta(GP_fp1["Sector2Time"])
GP_fp1['s3_TD']= pd.to_timedelta(GP_fp1["Sector3Time"])

GP_fp1.loc[:, "Time S1 (s)"] = GP_fp1["s1_TD"].dt.total_seconds()
GP_fp1.loc[:, "Time S2 (s)"] = GP_fp1["s2_TD"].dt.total_seconds()
GP_fp1.loc[:, "Time S3 (s)"] = GP_fp1["s3_TD"].dt.total_seconds()
```

```python
GP_fp1_Rk=GP_fp1[GP_fp1["LapTime"]!="NaT"]
GP_fp1_Rk["LAP_RK"] = GP_fp1_Rk.groupby("LapNumber")["LapTime (s)"].rank(method="dense", ascending=True)
GP_fp1_Rk["S1_RK"] = GP_fp1_Rk.groupby("LapNumber")["Time S1 (s)"].rank(method="dense", ascending=True)
GP_fp1_Rk["S2_RK"] = GP_fp1_Rk.groupby("LapNumber")["Time S2 (s)"].rank(method="dense", ascending=True)
GP_fp1_Rk["S3_RK"] = GP_fp1_Rk.groupby("LapNumber")["Time S3 (s)"].rank(method="dense", ascending=True)
GP_fp1_Rk["Race_RK"] = GP_fp1_Rk.groupby("GP")["LapTime (s)"].rank(method="dense", ascending=True)
```

```python
GP_fp1_Sectors=GP_fp1_Rk[["Driver","LapNumber","Time S1 (s)","Time S2 (s)","Time S3 (s)","LapTime","LapTime (s)","Compound","Year","GP","LAP_RK","S1_RK","S2_RK","S3_RK"]]
GP_fp1_Sectors=GP_fp1_Sectors.fillna(0)
GP_fp1_Sectors.sort_values(by=["LapNumber","S1_RK"]).head()
```

```python
#2024
def driver_color(driver):
    if driver =="VER":
        return "#3860C6"
    elif driver =="PER":
        return '#3860C6'
    elif driver =="NOR":
        return '#FF8000'
    elif driver =="PIA":
        return '#FF8000'
    elif driver =="LEC":
        return '#E80020'
    elif driver =="SAI":
        return '#E80020'
    elif driver =="HAM":
        return '#27F4D2'
    elif driver =="RUS":
        return '#27F4D2'
    elif driver =="ALO":
        return '#229971'
    elif driver =="STR":
        return '#229971'
    elif driver =="GAS":
        return '#0093cc'
    elif driver =="OCO":
        return '#0093cc'
    elif driver =="MAG":
        return '#B6BABD'
    elif driver =="HUL":
        return '#B6BABD'
    elif driver =="ZHO":
        return '#52e252'
    elif driver =="BOT":
        return '#52e252'
    elif driver =="ALB":
        return '#64C4FF'
    elif driver =="SAR":
        return '#64C4FF'
    elif driver =="COL":
        return '#64C4FF'    
    elif driver =="TSU":
        return '#6692FF'
    elif driver =="RIC":
        return '#6692FF'
    else:
        return '#3860C6'
```

```python
#get a combined view for fastest drivers in the entire lap, S1, S2, and S3
Fastest_Lap_fp1=GP_fp1_Sectors[GP_fp1_Sectors["LAP_RK"]==1][["GP","Year","LapNumber","LapTime","LapTime (s)","Compound","Driver"]].reset_index(drop=True)
Fastest_Lap_fp1.columns=["GP","Year","LapNumber","LapTime","LapTime (s)","Lap Compound","Fastest Lap"]
Fastest_sector1_fp1=GP_fp1_Sectors[GP_fp1_Sectors["S1_RK"]==1][["LapNumber","Time S1 (s)","Compound","Driver"]].reset_index(drop=True)
Fastest_sector1_fp1.columns=["LapNumber","Time S1 (s)","S1 Compound","Fastest S1"]
Fastest_sector2_fp1=GP_fp1_Sectors[GP_fp1_Sectors["S2_RK"]==1][["LapNumber","Time S2 (s)","Compound","Driver"]].reset_index(drop=True)
Fastest_sector2_fp1.columns=["LapNumber","Time S2 (s)","S2 Compound","Fastest S2"]
Fastest_sector3_fp1=GP_fp1_Sectors[GP_fp1_Sectors["S3_RK"]==1][["LapNumber","Time S3 (s)","Compound","Driver"]].reset_index(drop=True)
Fastest_sector3_fp1.columns=["LapNumber","Time S3 (s)","S3 Compound","Fastest S3"]
Fast_Laps_fp1=Fastest_Lap_fp1.merge(Fastest_sector1_fp1,on=['LapNumber']).merge(Fastest_sector2_fp1,on=['LapNumber']).merge(Fastest_sector3_fp1,on=['LapNumber'])
Fast_Laps_fp1=Fast_Laps_fp1[["GP","Year","LapNumber","LapTime","LapTime (s)","Fastest Lap","Lap Compound","Time S1 (s)","Fastest S1","S1 Compound","Time S2 (s)","Fastest S2","S2 Compound","Time S3 (s)","Fastest S3","S3 Compound"]].sort_values(by=["LapNumber"])
Fast_Laps_fp1=Fast_Laps_fp1.reset_index(drop=True)
Fast_Laps_fp1.head()
```

```python
Fast_Laps_fp1['color'] = Fast_Laps_fp1['Fastest Lap'].apply(driver_color)
Fast_Laps_fp1.head()
```

```python
LapTime_fp1_ordered=Fast_Laps_fp1.sort_values(by=['LapTime (s)'])
LapTime_fp1_ordered.head()
```

```python
#2024
race_multi=["Bahrain Grand Prix", "Saudi Arabian Grand Prix", "Australian Grand Prix", "Japanese Grand Prix","Chinese Grand Prix","Miami Grand Prix","Emilia Romagna Grand Prix","Monaco Grand Prix","Canadian Grand Prix","Spanish Grand Prix","Austrian Grand Prix","British Grand Prix","Hungarian Grand Prix","Belgian Grand Prix"]
#2023
# race_multi=["Abu Dhabi Grand Prix","Australian Grand Prix","Austrian Grand Prix","Azerbaijan Grand Prix","Bahrain Grand Prix","Belgian Grand Prix","British Grand Prix","Canadian Grand Prix","Dutch Grand Prix","Hungarian Grand Prix","Italian Grand Prix","Japanese Grand Prix","Las Vegas Grand Prix","Mexico City Grand Prix","Miami Grand Prix","Monaco Grand Prix","Pre-Season Testing","Qatar Grand Prix","Saudi Arabian Grand Prix","Singapore Grand Prix","Spanish Grand Prix","São Paulo Grand Prix","United States Grand Prix"]
# 2022
# race_multi=["Abu Dhabi Grand Prix","Australian Grand Prix","Austrian Grand Prix","Azerbaijan Grand Prix","Bahrain Grand Prix","Belgian Grand Prix","British Grand Prix","Canadian Grand Prix","Dutch Grand Prix","Emilia Romagna Grand Prix","French Grand Prix","Hungarian Grand Prix","Italian Grand Prix","Japanese Grand Prix","Mexico City Grand Prix","Miami Grand Prix","Monaco Grand Prix","Saudi Arabian Grand Prix","Singapore Grand Prix","Spanish Grand Prix","São Paulo Grand Prix","United States Grand Prix"]
#2021
# race_multi=["Abu Dhabi Grand Prix","Austrian Grand Prix","Azerbaijan Grand Prix","Bahrain Grand Prix","Belgian Grand Prix","British Grand Prix","Dutch Grand Prix","Emilia Romagna Grand Prix","French Grand Prix","Hungarian Grand Prix","Italian Grand Prix","Mexico City Grand Prix","Monaco Grand Prix","Portuguese Grand Prix","Qatar Grand Prix","Russian Grand Prix","Saudi Arabian Grand Prix","Spanish Grand Prix","Styrian Grand Prix","São Paulo Grand Prix","Turkish Grand Prix","United States Grand Prix"]
#2020
# race_multi=["70th Anniversary Grand Prix","Abu Dhabi Grand Prix","Austrian Grand Prix","Bahrain Grand Prix","Belgian Grand Prix","British Grand Prix","Eifel Grand Prix","Emilia Romagna Grand Prix","Hungarian Grand Prix","Italian Grand Prix","Portuguese Grand Prix","Russian Grand Prix","Sakhir Grand Prix","Spanish Grand Prix","Styrian Grand Prix","Turkish Grand Prix","Tuscan Grand Prix"]

year_multi=2024
for i in race_multi:
    df_bq_fp1=df_bq2_fp1[(df_bq2_fp1["GP"]==i)&(df_bq2_fp1["Year"]==year_multi)]
    GP_fp1=df_bq_fp1
    GP_fp1["LapNumber"]=GP_fp1["LapNumber"].astype(int)
    GP_fp1=GP_fp1.sort_values(['Driver','LapNumber'])
    GP_fp1["DriverNumber"]=GP_fp1["DriverNumber"].astype(int)
    # year=GP_fp1['Year'].iloc[0]
    # gp=GP_fp1['GP'].iloc[0]
    GP_fp1['LapTime_TD']= pd.to_timedelta(GP_fp1["LapTime"])
    GP_fp1["LapTime"]=GP_fp1['LapTime'].str.split('days ').str[1]
    GP_fp1.loc[:, "LapTime (s)"] = GP_fp1["LapTime_TD"].dt.total_seconds()
    GP_fp1['s1_TD']= pd.to_timedelta(GP_fp1["Sector1Time"])
    GP_fp1['s2_TD']= pd.to_timedelta(GP_fp1["Sector2Time"])
    GP_fp1['s3_TD']= pd.to_timedelta(GP_fp1["Sector3Time"])

    GP_fp1.loc[:, "Time S1 (s)"] = GP_fp1["s1_TD"].dt.total_seconds()
    GP_fp1.loc[:, "Time S2 (s)"] = GP_fp1["s2_TD"].dt.total_seconds()
    GP_fp1.loc[:, "Time S3 (s)"] = GP_fp1["s3_TD"].dt.total_seconds()
    GP_fp1_Rk=GP_fp1[GP_fp1["LapTime"]!="NaT"]
    GP_fp1_Rk["LAP_RK"] = GP_fp1_Rk.groupby("LapNumber")["LapTime (s)"].rank(method="dense", ascending=True)
    GP_fp1_Rk["S1_RK"] = GP_fp1_Rk.groupby("LapNumber")["Time S1 (s)"].rank(method="dense", ascending=True)
    GP_fp1_Rk["S2_RK"] = GP_fp1_Rk.groupby("LapNumber")["Time S2 (s)"].rank(method="dense", ascending=True)
    GP_fp1_Rk["S3_RK"] = GP_fp1_Rk.groupby("LapNumber")["Time S3 (s)"].rank(method="dense", ascending=True)
    GP_fp1_Rk["Race_RK"] = GP_fp1_Rk.groupby("GP")["LapTime (s)"].rank(method="dense", ascending=True)
    GP_fp1_Sectors=GP_fp1_Rk[["Driver","LapNumber","Time S1 (s)","Time S2 (s)","Time S3 (s)","LapTime","LapTime (s)","Compound","Year","GP","LAP_RK","S1_RK","S2_RK","S3_RK"]]
    GP_fp1_Sectors=GP_fp1_Sectors.fillna(0)
    Fastest_Lap_fp1=GP_fp1_Sectors[GP_fp1_Sectors["LAP_RK"]==1][["GP","Year","LapNumber","LapTime","LapTime (s)","Compound","Driver"]].reset_index(drop=True)
    Fastest_Lap_fp1.columns=["GP","Year","LapNumber","LapTime","LapTime (s)","Lap Compound","Fastest Lap"]
    Fastest_sector1_fp1=GP_fp1_Sectors[GP_fp1_Sectors["S1_RK"]==1][["LapNumber","Time S1 (s)","Compound","Driver"]].reset_index(drop=True)
    Fastest_sector1_fp1.columns=["LapNumber","Time S1 (s)","S1 Compound","Fastest S1"]
    Fastest_sector2_fp1=GP_fp1_Sectors[GP_fp1_Sectors["S2_RK"]==1][["LapNumber","Time S2 (s)","Compound","Driver"]].reset_index(drop=True)
    Fastest_sector2_fp1.columns=["LapNumber","Time S2 (s)","S2 Compound","Fastest S2"]
    Fastest_sector3_fp1=GP_fp1_Sectors[GP_fp1_Sectors["S3_RK"]==1][["LapNumber","Time S3 (s)","Compound","Driver"]].reset_index(drop=True)
    Fastest_sector3_fp1.columns=["LapNumber","Time S3 (s)","S3 Compound","Fastest S3"]
    Fast_Laps_fp1=Fastest_Lap_fp1.merge(Fastest_sector1_fp1,on=['LapNumber']).merge(Fastest_sector2_fp1,on=['LapNumber']).merge(Fastest_sector3_fp1,on=['LapNumber'])
    Fast_Laps_fp1=Fast_Laps_fp1[["GP","Year","LapNumber","LapTime","LapTime (s)","Fastest Lap","Lap Compound","Time S1 (s)","Fastest S1","S1 Compound","Time S2 (s)","Fastest S2","S2 Compound","Time S3 (s)","Fastest S3","S3 Compound"]].sort_values(by=["LapNumber"])
    Fast_Laps_fp1=Fast_Laps_fp1.reset_index(drop=True)
    Fast_Laps_fp1['color'] = Fast_Laps_fp1['Fastest Lap'].apply(driver_color)
    LapTime_fp1_ordered=Fast_Laps_fp1.sort_values(by=['LapTime (s)'])
    fig_fp1 = go.Figure(go.Table
    (header={"values": LapTime_fp1_ordered.columns},
    cells=dict(
        values=LapTime_fp1_ordered.T.values,
        fill_color=[LapTime_fp1_ordered.color],
        align='center', font=dict(color='white', size=11)
        ))
    )
    fig_fp1.update_layout(
        title="FP1 Fastest Lap and Sector Times Ordered By Lap Times For The {} {}".format(year_multi, i),
            title_x=0.5,
        yaxis = dict(tickfont = dict(size=15)),
        xaxis = dict(tickfont = dict(size=15)),
        font=dict(
            family="PT Sans Narrow",
            size=13,
            color="Black"
        ),
        height=780,
        width=1200,
        title_font_family="PT Sans Narrow",
        margin=dict(l=5, r=5, t=35, b=5),
    )
    fig_fp1.write_html("/Users/rdesh723/statpulse-html/plots/Free Practice/FP1/{}/{} {} FP1 Fastest Lap Times and Sectors Ordered by Lap Times.html".format(save_year_multi,year_multi, i),include_plotlyjs='cdn')
```
