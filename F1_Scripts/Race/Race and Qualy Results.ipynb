```python
#Change the date to the previous day
query = """
SELECT
*
FROM
`perceptive-ivy-290216.f1_api.results_race`
# WHERE Year=2024 AND GP="Austrian Grand Prix"
"""
project_id = "perceptive-ivy-290216"
df_bq2_race = pandas_gbq.read_gbq(query, project_id=project_id, dialect='standard')
```

```python
racing='Belgian Grand Prix'
```

```python
df_bq_race=df_bq2_race[(df_bq2_race["GP"]==racing)&(df_bq2_race["Year"]==2024)]
GP_race=df_bq_race
GP_race["DriverNumber"]=GP_race["DriverNumber"].astype(int)
year_race=GP_race['Year'].iloc[0]
gp_race=GP_race['GP'].iloc[0]
GP_race.head()
```

```python
def driver_color(driver):
    if driver =="VER":
        return "#3860C6"
    elif driver =="PER":
        return '#3860C6'
    elif driver =="NOR":
        return '#FF8000'
    elif driver =="PIA":
        return '#FF8000'
    elif driver =="LEC":
        return '#E80020'
    elif driver =="SAI":
        return '#E80020'
    elif driver =="HAM":
        return '#27F4D2'
    elif driver =="RUS":
        return '#27F4D2'
    elif driver =="ALO":
        return '#229971'
    elif driver =="STR":
        return '#229971'
    elif driver =="GAS":
        return '#0093cc'
    elif driver =="OCO":
        return '#0093cc'
    elif driver =="MAG":
        return '#B6BABD'
    elif driver =="HUL":
        return '#B6BABD'
    elif driver =="ZHO":
        return '#52e252'
    elif driver =="BOT":
        return '#52e252'
    elif driver =="ALB":
        return '#64C4FF'
    elif driver =="SAR":
        return '#64C4FF'
    elif driver =="TSU":
        return '#6692FF'
    elif driver =="RIC":
        return '#6692FF'
    else:
        return '#3860C6'
```

```python
GP_race['Color'] = GP_race['Abbreviation'].apply(driver_color)
GP_race.head()
```

```python
Fast_Laps_Race=GP_race[["Year","GP",'TeamName','FullName','GridPosition','Position','ClassifiedPosition','Status','Time','Points','Color']].sort_values(by=['Position'])
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Table
 (header={"values": Fast_Laps_Race.columns},
  cells=dict(
    values=Fast_Laps_Race.T.values,
    fill_color=[Fast_Laps_Race.Color],
    align='center', font=dict(color='white', size=12)
    ))
 )

# fig.update_layout(
#     updatemenus=[
#         {
#             "buttons": [
#                 {
#                     "label": c,
#                     "method": "update",
#                     "args": [
#                         {
#                             "cells": {
#                                 "values": Fast_Laps.T.values
#                                 if c == "All"
#                                 else Fast_Laps.loc[Fast_Laps["TeamName"].eq(c)].T.values
#                             }
#                         }
#                     ],
#                 }
#                 for c in ["All"] + Fast_Laps["TeamName"].unique().tolist()
#             ]
#         }
#     ]
# )

fig.update_layout(
    title="{} {} Race Results".format(year_race,gp_race),
        title_x=0.5,
        # template="plotly_dark",

    yaxis = dict(tickfont = dict(size=20)),
    xaxis = dict(tickfont = dict(size=20)),
    font=dict(
        family="PT Sans Narrow",
        size=15,
        color="Black"
    ),
    height=1000,
    width=1050,
    title_font_family="PT Sans Narrow",
    margin=dict(l=5, r=5, t=35, b=5),
)

fig.show()
```

```python
fig.write_html("{} {} Race Results.html".format(year_race,gp_race))
```
